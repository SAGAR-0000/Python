```python
import matplotlib.pyplot as plt
import numpy as np

a = [50, 60, 70, 80, 90]    
b = [55, 65, 75, 90, 100]    

x = np.linspace(1, 51, 5)  

plt.bar(x, a, width=3, color="red", label="Australia")
plt.bar(x + 3, b, width=3, color="green", label="India")

plt.xlabel("Overs")
plt.ylabel("Runs Scored")
plt.title("Scoring Chart: India vs Australia")
plt.legend()
plt.show()
```

```python
from sklearn.impute import SimpleImputer
num_imputer=SimpleImputer(strategy="median")
cat_imputer=SimpleImputer(strategy="most_frequent")
```

```python
df1[numerical_cols]=num_imputer.fit_transform(df1[numerical_cols])
df1[categorical_cols]=cat_imputer.fit_transform(df1[categorical_cols])
```
