```python
import matplotlib.pyplot as plt 
import seaborn as sns

plt.rcParams["figure.figsize"]=[8,6]
sns.set_style("darkgrid")
titanic_data = sns.load_dataset("titanic")
titanic_data.head()
```

```python
td_c=titanic_data
```

```python
td_c
```

```python
td_c1=td_c[["sex","class","embark_town"]]
```

```python
td_c1
```

```python
print(td_c1["sex"].unique())
print(td_c1["class"].unique())
print(td_c1["embark_town"].unique())
```

```python
td_c1["embark_town"].fillna("Missing",inplace=True)
td_c1
```

```python
td_c1.isnull()
```

```python
import pandas as pd
temp=pd.get_dummies(td_c1["sex"], dtype=int)
temp.head()
```

```python
import pandas as pd
temp=pd.get_dummies(td_c1["class"], dtype=int)
temp.head()
```

```python
import pandas as pd
temp=pd.get_dummies(td_c1["embark_town"], dtype=int)
temp.head()
```

```python
pd.concat([td_c1["embark_town"], pd.get_dummies(td_c1["embark_town"], dtype=int)], axis=1).head()
```

```python
Label Encoding :
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(td_c1["class"])
td_c1["le_class"]=le.transform(td_c1["class"])
td_c1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(td_c1["sex"])
td_c1["le_sex"]=le.transform(td_c1["sex"])
td_c1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
le.fit(td_c1["embark_town"])
td_c1["le_embark_town"]=le.transform(td_c1["embark_town"])
td_c1.head()
```

```python
Frequency Encoding: 
```

```python
No of times a value appeared (not float/avg)
```

```python
value_counts=td_c1["embark_town"].value_counts().to_dict()
print(value_counts)
```

```python
td_c1["embark_town_frequency"]=td_c1["embark_town"].map(value_counts)
td_c1.head()
```

```python
del td_c1['embark_class']
```

```python
td_c1.head()
```

```python
value_counts=td_c1["class"].value_counts().to_dict()
print(value_counts)
td_c1["class_frequency"]=td_c1["class"].map(value_counts)
td_c1.head()
```

```python
value_counts=td_c1["sex"].value_counts().to_dict()
print(value_counts)
td_c1["sex_frequency"]=td_c1["sex"].map(value_counts)
td_c1.head()
```
