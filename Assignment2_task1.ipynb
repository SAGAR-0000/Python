```python
#Load a dataset into a pandas DataFrame and display the first few rows.
```

```python
import pandas as pd
df = pd.read_csv('Cost_of_Living_Index_by_Country_2024.csv')
print(df)
```

```python
df.head()
```

```python
#Filter the DataFrame to display only the rows that meet a certain condition
```

```python
print(df[df['Cost of Living Index']>50])
```

```python
#Sort the DataFrame based on a particular column in ascending or descending order.
```

```python
print(df.sort_values(by='Rent Index',ascending=False))
```

```python
#Add a new calculated column to the DataFrame based on existing columns
```

```python
df['Total Index'] = df['Cost of Living Index'] + df['Rent Index']
print(df)
```

```python
#Load two related datasets into separate DataFrames.
```

```python
import pandas as pd
df2 = pd.read_csv('countries.csv')
print(df2)
```

```python
#Perform an inner join on the two DataFrames based on a common key.
```

```python
inner_join = pd.merge(df, df2, on='Country', how='inner')
print(inner_join)
```

```python
#Perform an inner join on the two DataFrames based on a common key.
```

```python
left_join = pd.merge(df, df2, on='Country', how='left')
print(left_join)
```

```python
#Merge the two DataFrames using a concatenation operation and describe a scenario where this might be useful.
```

```python
concat_df = pd.concat([df, df], ignore_index=True)
print(concat_df.head())
```
