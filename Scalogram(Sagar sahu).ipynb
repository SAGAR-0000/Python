```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pywt

df = pd.read_csv(r"C:/Users/Asus/Desktop/iiitnr/synthetic_ppg_hr_hrv.csv")

signal = df.iloc[0, 0:1000].values  

scales = np.arange(1, 128)  
coefficients, freqs = pywt.cwt(signal, scales, 'morl')  

#Scalegram
plt.figure(figsize=(12, 6))
plt.imshow(np.abs(coefficients), extent=[0, 1000, 1, 128], aspect='auto',vmax=abs(coefficients).max(), vmin=0)
plt.colorbar(label='Power')
plt.xlabel("Time (samples)")
plt.ylabel("Scale")
plt.title("Sagar Sahu")
plt.show()

plt.savefig("Scalegram.png") 
```
